# Primer punto
Asignación de asientos en un avion de forma peculiar:


*   Los pasajeros ingresan al avión en el orden de su número de asiento (1,2,3...100).
*   El primer pasajero, en lugar de sentarse en su asiento asignado, elige uno al azar entre los 100 asientos disponibles.
*   Los demás pasajeros, al abordar, intentan sentarse en su asiento asignado. Sin embargo, si encuentran que su asiento está ocupado, deben elegir otro asiento aleatoriamente entre los que aún están disponibles.

**Pregunta:** ¿Cuál es la probabilidad de que el pasajero número 100 logre sentarse en su asiento originalmente asignado (el asiento número 100)?


```python
import random

# Número de iteraciones para simular
num_simulaciones = 100000
exitos = 0  # Contador de casos en que el pasajero 100 se sienta en su asiento

for _ in range(num_simulaciones):
    # Lista de asientos disponibles
    asientos_disponibles = list(range(1, 101))
    # El primer pasajero elige un asiento aleatorio
    asiento_primero = random.choice(asientos_disponibles)
    asientos_disponibles.remove(asiento_primero)
    # Los pasajeros 2 al 99 intentan sentarse en su asiento asignado
    pasajero = 2
    while pasajero < 100:
        if pasajero in asientos_disponibles:
            asientos_disponibles.remove(pasajero)
        else:
            asiento_random = random.choice(asientos_disponibles)
            asientos_disponibles.remove(asiento_random)
        pasajero += 1
    # Verificar si el pasajero 100 puede sentarse en su asiento
    if 100 in asientos_disponibles:
        exitos += 1

# Calcular la probabilidad
probabilidad = exitos/num_simulaciones
print(f"La probabilidad de que el pasajero número 100 se siente en su asiento es: {(probabilidad)*100}%")
```

# Segundo Punto
Estimción de $\pi$ mediante el método de **Monte Carlo**

image.png

Pasos básicos:
1.   Simula un cuadrado y un círculo y generar puntos aleatorios $(x, y)$ dentro del cuadrado $[-0.5, 0.5] \times [-0.5, 0.5]$.
2.   Determinar si cada punto está dentro del círculo con la ecuación $x^2 + y^2 \leq r^2$, donde el radio $r = 0.5$.
3.   Calcular la proporción de puntos dentro del círculo respecto al total de puntos generados.
4.   Usar la relación área círculo/cuadrado para aproximar $\pi$.


```python
import random

num_puntos = 50000000
cont_pnts = 0

# Generar puntos y verificar si caen dentro del círculo
for _ in range(num_puntos):
    x = random.uniform(-0.5, 0.5)
    y = random.uniform(-0.5, 0.5)

    # Verificar si el punto está dentro del círculo
    if x**2 + y**2 <= 0.5**2:  # Radio del círculo es 0.5
        cont_pnts += 1

# Calcular la estimación de pi
pi_estimado = 4 * (cont_pnts/num_puntos)

# Mostrar el resultado
print(f"El valor estimado de π usando el método de Monte Carlo es: {pi_estimado:.5f}")
```
