# **Taller 4**: Numpy
Implementación de regresión lineal usando la fórmula matricial en Numpy
$$ \beta=(X^{T}X)^{-1}X^{T} $$


```python
import numpy as np
import matplotlib.pyplot as plt
## Puntos
x=np.array([1,2,3,4,5])
y=np.array([1,3,6,7,10])

plt.scatter(x,y)
plt.title("Datos de estudio")
plt.xlabel("Horas de estudio")
plt.ylabel("Calificación")
```

```python
## Regresión Lineal
#Matriz de diseño
X=np.column_stack((np.ones(len(x)),x))
print(X)
```

```python
# Calculamos los betas de la regresión
p1=np.dot((np.transpose(X)),X)
p2=np.linalg.inv(p1)
p3=np.dot(p2,np.transpose(X))
Beta=np.dot(p3,y)
print(f"Betas: {Beta}")
Intercepto=Beta[0] ## b
Coeficiente=Beta[1] ## m
y_predct=np.dot(X,Beta)
print(f"Predicción de y: {y_predct}")

plt.scatter(x,y, label="Datos")
plt.plot(x,y_predct, color="red", label=f"Regresión Lineal: y={Coeficiente:.2f} x + {Intercepto:.2f}")
plt.xlabel("Horas de estudio")
plt.ylabel("Calificación")
plt.title("Regresión Lineal")
plt.legend()
plt.show()

```
